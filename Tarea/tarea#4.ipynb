```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine=datasets.load_wine()
wine
```

```python
wine_dataframe=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_dataframe
```

## Base de datos Wine 
Estos datos son el resultado de un anÃ¡lisis quÃ­mico de vinos cultivados en la misma regiÃ³n en Italia pero derivados de tres cultivares diferentes. El anÃ¡lisis determinÃ³ las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos.

```python
wine_dataframe["target"]=wine.target
wine_dataframe
x=wine_dataframe["alcohol"]
print(x.mean())
y=wine_dataframe["malic_acid"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
alcohol=wine_dataframe["alcohol"]*50
malic_acid=wine_dataframe["malic_acid"]
```

```python
plt.scatter(x,y,c=col,s=alcohol,alpha=0.5)
plt.xlabel("alcohol")
plt.ylabel("malic_acid")
plt.title("alcohol vs malic_acid")
plt.grid(True)
plt.show()
```
