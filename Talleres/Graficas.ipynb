```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color="brown",linewidth=5,alpha=0.5, marker="o", markersize=15,markeredgecolor="g",markerfacecolor="magenta",markeredgewidth=1)
plt.xlabel("$x(cm^2)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento.1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.title("funcion Sen(4x)0-$2\pi$"),
plt.grid(True)
plt.plot(x,y,color="Green")
plt.plot(x,y2,color="darkslategray", dashes=[8,2,2,2])
plt.xlabel("t(seg)")
plt.ylabel("Amplitud de la onda")
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*150+150
plt.scatter(x,y,s=r,color="black")
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50, facecolor="slategray",edgecolor="black",density=(True))
plt.title("distribucion de la poblacion")
plt.ylabel("Densidad de habitantes/$m$")
plt.text(-3,.35,r"$\mu=100,\ \sigma=1.0$")
plt.show()
```
